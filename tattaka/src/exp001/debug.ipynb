```python
import matplotlib.pyplot as plt
import numpy as np

def dice(mask1, mask2, use_mask):
    mask1 = mask1 * use_mask
    mask2 = mask2 * use_mask
    intersection = 2.0 * (mask1 * mask2).sum()
    union = mask1.sum() + mask2.sum()
    if mask1.sum() == 0 and mask2.sum() == 0:
        return 1.0

    return intersection / union

y_valid = np.load("../../logs/exp000/resnetrs50_baseline/fold0/debug_y_valid.npy")
p_valid = np.load("../../logs/exp000/resnetrs50_baseline/fold0/debug_p_valid.npy")
plt.imshow(y_valid)
plt.show()
plt.imshow(p_valid)
dice(y_valid, p_valid, use_mask=np.ones_like(p_valid))
```
